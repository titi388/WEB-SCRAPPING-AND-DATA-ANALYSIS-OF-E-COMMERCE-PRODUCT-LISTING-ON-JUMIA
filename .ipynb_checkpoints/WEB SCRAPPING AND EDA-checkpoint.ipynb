```python
# The necessary package
import selenium
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup as bs
```

```python
chromedriver = "C:/Users/HP/Desktop/chromedriver/chromedriver.exe"

opt = Options()
opt.add_argument("--headless")
svr =  Service(chromedriver)
driver = webdriver.Chrome(options=opt, service=svr)
```

```python
url = "https://www.amazon.com/s/ref=nb_sb_noss?url=search-alias%3Daps&field-keywords=watch"
driver.get(url) # getting the access to the web page
driver.implicitly_wait(10) # wait time on the web page
html_content = driver.page_source
html_content
```

```python
soup = bs(html_content)
```

```python
container=soup.findAll("div", {"class": "a-section a-spacing-base"})
```

```python
name = []
price = []
for i in container:
    try:
        name.append(i.find("span", class_= "a-size-base-plus a-color-base a-text-normal").text)

    except:
        name.append("no name")
    try:
        price.append(i.find ("span", class_= "a-price-whole").text)
    except:
        price.append("no price")
```

```python
import pandas as pd
d = {"watch name": name, "watch price":price}
y=pd.DataFrame(d)
```
